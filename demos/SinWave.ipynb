# A sine wave

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets
```

The basic functional form for a sine wave is:

$$y(t) = A \sin(\omega t)$$

where:
* A - amplitude
* $\omega$ - frequency
* t - time
* y - wave amplitude as a function of time

```python
from IPython.display import HTML
HTML('<img src="gifs/harm_osc.gif" width=200>')
```

```python
def plotsin(A=1,w=1):
    t = np.linspace(0, 2*np.pi, 1000)
    y = A*np.sin(w*t)
    fig,ax=plt.subplots()
    plt.plot(t,y)
    plt.xlim(0, 2*np.pi)
    plt.ylim(-20, 20)
    plt.xlabel('t',fontsize=14)
    plt.ylabel('y',fontsize=14)
    ax.text(7, 6, f'We\'re plotting the equation: $y(t) = {A} \sin({w} t)$', fontsize=12)


ipywidgets.interact(plotsin, A=(0,20), w=5);
```

```python
def plotsin(A=1,k=1):
    t = np.linspace(0, 2*np.pi, 1000)
    y = A*np.sin(k*t)
    y2 = np.sin(t)
    fig,ax=plt.subplots()
    plt.plot(t,y,t,y2)
    plt.ylim(-1.5,1.5)
    plt.xlim(0, 2*np.pi)
    plt.xlabel('t',fontsize=14)
    plt.ylabel('y',fontsize=14)
    plt.grid(True)
    ax.text(1.2, 1.1, f'$y(t) = \sin(t)$', fontsize=12)
    ax.text(0.75, -0.45, f'$y(t) = A$ $\sin(\omega t)$', fontsize=12)
    ax.text(0.75, -0.7, f'A=0.25, $\omega$=2', fontsize=12)
    plt.savefig('sinipy.png')

plotsin(A=.25, k=2);
```
